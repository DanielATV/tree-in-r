```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('smogon.csv',sep=',')
```

```python
df.head()
```

```python
df.replace({
    'BL':'OU','BL2':'UU','BL3':'RU','BL4':'NU','AG':'Uber'
},inplace=True)
```

```python
df['Type.2'].fillna('NoType',inplace=True)

df[df['Type.2'].isna()]['Type.2'] = 'NoType'
```

```python
df.head()
```

```python
df.columns = ['DexNum','Name','Type1',
                     'Type2','Total','HP','Atk','Def','SAtk','SDef','Spd','Generation','Legendary','Mega','Tier']
```

```python
df.head()
```

```python
df.replace({"Farfetch'd":'Farfetchd'},inplace=True)
```

```python
df.info()
```

```python
df.to_csv('smogon_cleaned.csv',sep=';',index=False)
```
